```python
def pie_slice(ep, R, theta1, theta2):
    return [
        lambda t: exp(i * theta1) * (ep + (R - ep) * t),
        lambda t: exp(i * (theta1 + (theta2 - theta1) * t)) * R,
        lambda t: exp(i * theta2) * (R + (ep - R) * t),
        lambda t: exp(i * (theta2 + (theta1 - theta2) * t)) * ep,
    ]
```

```python
def log_cut(z, theta):
    theta = theta + pi 
    return theta * i + log(z * exp(- i * theta))
```

```python
def complex_parametric(f):
    time = ellipsis_range(0, 1/100, Ellipsis, 1)
    return line([f(t) for t in time])
    
show(sum([complex_parametric(p)for p in pie_slice(.1, 4, 0, pi / 4)]))
    
```

```python
gamma1 = list(pie_slice(.1, 4., 0., 3 * pi / 4))
integrand_1(x) = log_cut(x, -.01 * pi / 4) / (1 + x^2)
bg= complex_plot(integrand_1, (-4, 4), (-4, 4))

show(bg + sum([complex_parametric(p)for p in gamma1]))

sum([integral(integrand_1(p(t)) * derivative(p(t), t), t, 0, 1) for p in gamma1])
```
