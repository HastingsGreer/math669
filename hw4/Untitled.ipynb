```python
import numpy as np
```

```python
z = np.mgrid[-4:4:100j, -4:4:100j]
z = z[0] * 1j + z[1]
```

```python
z

def x(R, theta):
    return R * np.exp( 1j * theta)
```

```python
def heightmap(theta):
    return np.real(z * x(1, theta) - z**3 / 3)
```

```python
import matplotlib.pyplot as plt
ext= (-4, 4, -4, 4)
plt.contour(heightmap(0), levels=80, cmap="jet", extent=ext)

plt.colorbar()
```
